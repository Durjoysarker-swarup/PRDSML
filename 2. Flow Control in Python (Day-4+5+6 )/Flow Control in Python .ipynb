## ifâ€¦elseâ€¦ Statements

```python
name = input("What is your name? ")
passward = "swardfish"
if name == "Mary":
    print("Hello, Mary")
    if passward == "swordfish":
        print("Access granted.")
    else:
        print("Wrong passward.")
else:
    print("Hello, stranger.")
```

## elif Statements

```python
name = 'Carol'
age = 3000
if name == 'Alice':
    print('Hi, Alice.')
elif age < 12:
    print('You are not Alice, kiddo.')
elif age > 2000:
    print('Unlike you, Alice is not an undead, immortal vampire.')
elif age > 100:
    print('You are not Alice, grannie.')
```

```python
name = 'Carol'
age = 30
if name == 'Alice':
    print('Hi, Alice.')
elif age < 12:
    print('You are not Alice, kiddo.')
else:
    print('You are neither Alice nor a little kid.')
```

## While loop statement

```python
spam = 0
if spam < 5:
    print('Hello, world.')
    spam = spam + 1
    print(spam)
```

- while ->Run the until the case is false.

```python
spam = 0
while spam < 5:
    print('Hello, world.')
    spam = spam + 1
    print(spam)
```

## break statement

!= -> not equal to

```python
name = ""
while name != "your name":
    print("Please type your name.")
    name = input()
print("Thank you!")
```

```python
while True:
    name = input("who are you? ")
    if name == "your name":
      break
print("Thank you!")
```

## continue Statement

```python
while True:
  name = input("Who are you? ")
  if name != "Joe":
    continue
  password = input("Hello, Joe. What is the password? (It is a fish.) ")
  if password == "swordfish":
    break
print("Access granted.")
```
