### 2 parameter

```python
def hello(name, age):
  print(f"Hello {name} !")
  print(f"You are {age} years old.")
```

```python
hello("Durjoy", 24)
```

## The call stack

```python
def a():
  print("a() starts")
  b()
  d()
  print("a() ends")

def b():
  print("b() starts")
  c()
  print("b() ends")

def c():
  print("c() starts")
  print("c() ends")

def d():
  print("d() starts")
  print("d() ends")
```

```python
a()
```

## The return statement

```python
def add(a,b):
  return f"the result is {a+b}"
```

```python
add(4,5)
```

## BMI Calculations

```python
def bmi_calculator():
  name = input("Enter your name: ")
  weight_kg = float(input("Enter your weight in kg: "))
  height_inch = float(input("Enter your height in inch: "))
  height_m = height_inch * 0.0254
  bmi = weight_kg / (height_m ** 2)
  return f"{name}'s BMI is {bmi}"
```

```python
bmi_calculator()
```
